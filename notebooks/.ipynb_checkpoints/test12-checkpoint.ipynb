```python
#肺の画像認識
#COVID-19か正常かを分類するモデルを作成
import os
list_helthy = [i for i in os.listdir('data/images/COVID-NORMAL/healthy') if not i.startswith('.')]
list_covid = [i for i in os.listdir('data/images/COVID-NORMAL/covid19') if not i.startswith('.')]
```

```python
#画像ファイル集計
num_helthy = len(list_helthy)
num_covid = len(list_covid)
num_all = num_helthy + num_covid
print("正常画像ファイル数:", num_helthy)
print("COVID-19画像ファイル数:", num_covid)
print("総ファイル数", num_all)
```

```python
#画像データ格納用numpy配列作成
import numpy as np
image_temp = np.zeros((num_all, 64, 64, 1))
labels_temp = np.zeros((num_all, 1),dtype = int)
print(image_temp.shape)
print(labels_temp.shape)
```

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
#健康画像の読み込み
path_helthy = 'data/images/COVID-NORMAL/healthy/'
for i in range(num_helthy):
    file = f'{path_helthy}/{list_helthy[i]}'
    file_image = load_img(
        file, 
        color_mode='grayscale', 
        target_size=(64,64),
        interpolation='lanczos'
        )
    image_temp[i] = img_to_array(file_image)/255
    labels_temp[i] = 0  #正常ラベル0
#COVID-19画像の読み込み
path_covid = 'data/images/COVID-NORMAL/covid19/'
for i in range(num_covid):
    file = f'{path_covid}/{list_covid[i]}'
    file_image = load_img(
        file, 
        color_mode='grayscale', 
        target_size=(64,64),
        interpolation='lanczos'
        )
    image_temp[num_helthy + i] = img_to_array(file_image)/255
    labels_temp[num_helthy + i] = 1  #COVID-19ラベル1

#順番をシャッフルする #COVID-19画像が後ろに偏っているため
num_list = np.arange(num_all)
np.random.seed(111)
np.random.shuffle(num_list)
x_train = image_temp[num_list]
y_train = labels_temp[num_list]
```

```python
#画像予測用深層学習モデルの作成
from keras.models import Sequential
from keras.layers import Dense, Flatten

model = Sequential()
model.add(Flatten(input_shape=(64,64,1)))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
result = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(result.history['accuracy'], label='accuracy')
plt.plot(result.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()
plt.plot(result.history['loss'], label='loss')
plt.plot(result.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#Dropout層を追加したモデルの作成
from keras.layers import Dropout

model_2 = Sequential()
model_2.add(Flatten(input_shape=(64,64,1)))
model_2.add(Dense(512, activation='relu'))
model_2.add(Dropout(0.5))
model_2.add(Dense(256, activation='relu'))
model_2.add(Dropout(0.5))
model_2.add(Dense(128, activation='relu'))
model_2.add(Dropout(0.5))
model_2.add(Dense(1, activation='sigmoid'))
model_2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_2.summary()   
```

```python
result_2=model_2.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
plt.plot(result_2.history['accuracy'], label='accuracy')
plt.plot(result_2.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()
plt.plot(result_2.history['loss'], label='loss')
plt.plot(result_2.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#未知画像で予測の実行
img1 = img_to_array(load_img(
    'data/images/covid.jpg',
    color_mode='grayscale',
    target_size=(64,64)))/255
img2 = img_to_array(load_img(
    'data/images/NORMAL.jpg',
    color_mode='grayscale',
    target_size=(64,64)))/255

#予測用配列作成
x_pred = np.zeros((2,64,64,1))
x_pred[0] = img1
x_pred[1] = img2
pred = model_2.predict(x_pred)
for i in range(len(pred)):
    if pred[i] < 0.5:
        print(f'画像{i+1}は正常画像です。確率:{1 - pred[i][0]:.2f}')
    else:
        print(f'画像{i+1}はCOVID-19画像です。確率:{pred[i][0]:.2f}')
print(pred)
```
